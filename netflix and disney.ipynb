```python

import unicodedata

# Ruta a los archivos CSV con caracteres incorrectos
netflix_file_path = 'D:/cursos/challenge rockingdata/netflix_titles.csv'
disney_file_path = 'D:/cursos/challenge rockingdata/disney_plus_titles.csv'

# Función para corregir caracteres UTF-8 mal interpretados
def corregir_caracteres(texto):
    try:
        texto = unicodedata.normalize('NFD', texto).encode('ascii', 'ignore').decode('utf-8')
        return texto
    except Exception as e:
        return texto

# Función para corregir caracteres en un archivo y guardar los datos corregidos
def corregir_caracteres_en_archivo(archivo_entrada, archivo_salida):
    data = pd.read_csv(archivo_entrada, sep=',')
    data = data.applymap(corregir_caracteres)
    data.to_csv(archivo_salida, index=False, sep=',', encoding='utf-8')
    return data

# Corregir caracteres en el archivo de Netflix y guardar los datos corregidos
netflix_data_corregido = corregir_caracteres_en_archivo(netflix_file_path, 'D:/cursos/challenge rockingdata/netflix_titles_corregidos.csv')

# Corregir caracteres en el archivo de Disney y guardar los datos corregidos
disney_data_corregido = corregir_caracteres_en_archivo(disney_file_path, 'D:/cursos/challenge rockingdata/disney_plus_titles_corregidos.csv')

# Muestra de las primeras filas de los datos corregidos
netflix_data_corregido.head()
disney_data_corregido.head()

```

```python

# Rutas de los archivos corregidos
netflix_file_path = 'D:/cursos/challenge rockingdata/netflix_titles_corregidos.csv'
disney_file_path = 'D:/cursos/challenge rockingdata/disney_plus_titles_corregidos.csv'

# Cargar los datos corregidos de Netflix y Disney
netflix_data = pd.read_csv(netflix_file_path, sep=',')
disney_data = pd.read_csv(disney_file_path, sep=',')

# Agregar una columna 'plataforma' con el identificador numérico de plataforma
netflix_data['plataforma'] = 1  # 1 representa Netflix
disney_data['plataforma'] = 2  # 2 representa Disney+

# Reorganizar el orden de las columnas para que 'plataforma' sea la primera
netflix_data = netflix_data[['plataforma'] + [col for col in netflix_data.columns if col != 'plataforma']]
disney_data = disney_data[['plataforma'] + [col for col in disney_data.columns if col != 'plataforma']]

# Concatenar los DataFrames de Netflix y Disney
datos_combinados = pd.concat([netflix_data, disney_data], ignore_index=True)

# Convertir la columna 'date_added' a tipo datetime
datos_combinados['date_added'] = pd.to_datetime(datos_combinados['date_added'], errors='coerce')

# Guardar los datos combinados en un nuevo archivo CSV
output_combined_file_path = 'D:/cursos/challenge rockingdata/netflix_disney_combined.csv'
datos_combinados.to_csv(output_combined_file_path, index=False, sep=';', encoding='utf-8')  # Guardar como CSV con comas como delimitador y codificación UTF-8

# Muestra de las primeras filas de los datos combinados
datos_combinados.head()
```

```python
from importdb import PostgreSQLConnection
from datetime import datetime

csv_file = 'netflix_disney_combined_continuous_show_id.csv'

data = pd.read_csv(csv_file, sep=';')

db_connection = PostgreSQLConnection()

# Mapeo de tipo de show a valores numéricos
showtype_mapping = {
    'Movie': 1,
    'TV Show': 2
}

# Convertir el campo 'type' en 'id_showtype'
data['id_showtype'] = data['type'].map(showtype_mapping)

# Agregar una nueva columna 'duration_str' que contiene "min" o "season"
data['duration_str'] = data['duration'].str.extract(r'(\D+)')

# Seleccionar las columnas necesarias
selected_columns = [
    'plataforma', 'id_showtype', 'title', 'date_added', 'duration', 'duration_str',
    'release_year', 'listed_in', 'description'
]

# Renombrar las columnas para que coincidan con la base de datos
data = data[selected_columns]
data.columns = [
    'id_platform', 'id_showtype', 'title', 'date_added', 'duration', 'duration_str',
    'release_year', 'listed_in', 'description'
]

# Modificar el campo 'duration' para extraer el número
data['duration'] = data['duration'].str.extract(r'(\d+)').fillna(0).astype(int)

# Convertir el campo 'date_added' al formato adecuado
def parse_date(date_str):
    if pd.isna(date_str):
        return None
    formats = ["%d/%m/%Y", "%B %d, Y"]
    for date_format in formats:
        try:
            return datetime.strptime(date_str, date_format).strftime("%Y-%m-%d")
        except ValueError:
            continue
    return None

data['date_added'] = data['date_added'].apply(lambda x: parse_date(x) if isinstance(x, str) else None)

# Insertar los datos en la tabla 'Shows'
db_connection.save_data_to_postgres('shows', data)
```

```python
from importdb import PostgreSQLConnection

csv_file = 'netflix_disney_combined_continuous_show_id.csv'

data = pd.read_csv(csv_file, sep=';')

# Crea una lista para almacenar los países únicos
unique_countries = []

# Divide el campo "country" en celdas separadas para manejar múltiples países
for countries in data['country']:
    if isinstance(countries, str):
        countries = countries.split(',')
        for country in countries:
            cleaned_country = country.strip()  # Limpia el nombre del país
            if cleaned_country not in unique_countries:
                unique_countries.append(cleaned_country)

db_connection = PostgreSQLConnection()

# Crea un DataFrame con id_country serial y country_name
country_data = pd.DataFrame({'country_name': unique_countries})

# Inserta los datos en la tabla 'country' (si no existen)
db_connection.save_data_to_postgres('country', country_data)

print("Operación completada.")
```
